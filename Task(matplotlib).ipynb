```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
path="/content/drive/MyDrive/tip.csv"
df=pd.read_csv(path)
```

```python
print("(Rows,Columns)=",df.shape)
print("Info",df.info())
print()
print("Null values :")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
plt.plot(df.tip,df.day)
plt.show()
```

```python
plt.bar(df.sex,df.total_bill)
plt.title("Total_bill vs sex")
plt.xlabel("Sex")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.hist(df.total_bill,bins=10,color="orange",edgecolor="red")
plt.title("Total Bill Histogram")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df.total_bill,df.tip)
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.boxplot(df.total_bill)
plt.title("Total Bill Boxplot")
plt.xlabel("Total Bill")
plt.show()
```

```python
x=df.groupby('day')['total_bill'].sum()
print(x)
plt.pie(x,labels=x.index,autopct='%1.1f%%')
plt.title("Total Bill by Day")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(df['size'],df['total_bill'])
plt.title("Total_bill vs size")
plt.xlabel("Size")
plt.ylabel("Total Bill")
plt.subplot(1,2,2)
plt.scatter(df.total_bill,df.day)
plt.title("Total Bill vs day")
plt.xlabel("Total Bill")
plt.ylabel("Day")
plt.show()
```
