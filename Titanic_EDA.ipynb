```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
df.isnull().sum()
```

```python
missing=pd.DataFrame({
    'missing_data':df.isnull().sum(),
    'missing_value':df.isnull().sum()/len(df)*100,
})
missing.sort_values('missing_value',ascending=False)
```

```python
df['age']=df['age'].fillna(df['age'].mean())
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0])
df.drop(columns=['deck'],inplace=True)
print("missing values after clean ", df.isnull().sum().sum())
```

```python
print(df.head())
```

```python
missing=pd.DataFrame({
    'missing_data':df.isnull().sum(),
    'missing_value':df.isnull().sum()/len(df)*100,
})
missing.sort_values('missing_value',ascending=False)
```

```python
print("duplicated rows: ", df.duplicated().sum())
df['sex']=df['sex'].astype('category')
df['embarked']=df['embarked'].astype('category')
df['class']=df['class'].astype('category')
df.info()
```

```python
num_cols=df.select_dtypes(include=["int64","float64"]).columns
df[num_cols].describe()
```

```python
sns.histplot(df['age'],kde=True,bins=40,color='orange',edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age distribution of passengers")
plt.show()
```

```python
sns.boxplot(df['fare'])
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Fare distribution of outlier")
plt.show()
```

```python
cat_col=df.select_dtypes(include=['category']).columns
for col in cat_col:
  sns.catplot(x=col,data=df)
  plt.title(f"count of {col}")
  plt.show()
```

```python
# numerical vs categorical(bi varuiate)
sns.boxplot(x="class",y="fare",data=df)
plt.xlabel("class")
plt.ylabel("fare")
plt.title("class vs fare")
plt.show()


sns.boxplot(x="sex",y="age",data=df)
plt.xlabel("sex")
plt.ylabel("age")
plt.title("sex vs age")
plt.show()
```

```python
# categorical vs categorical (bi avriate)
sns.boxplot(x="survived",hue="sex",data=df)
plt.title("survived vs sex")
plt.show()
```

```python
# categorical vs categorical (bi avriate)
sns.boxplot(x="survived",hue="class",data=df)
plt.title("survived vs class")
plt.show()
```

```python
# correlation and multivariate analysis
corr=df[["age","fare","survived"]].corr()
sns.heatmap(corr,annot=True,cbar=False)
plt.title("correlation")
plt.show()




sns.pairplot(df[["age","fare","survived"]],hue="survived")
plt.show()
```

```python
# outlier detection using iqr
q1=df['fare'].quantile(0.25)
q3=df['fare'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliers=df[(df['fare']<lower_bound) | (df['fare']>upper_bound)]
print("no of outliers: ", len(outliers))

df['fare_capped']=df['fare'].clip(lower=lower_bound,upper=upper_bound)
sns.boxplot(df['fare_capped'])
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Fare distribution of outlier")
```
