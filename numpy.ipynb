Given data (raw)

production â€” shape (3 farms Ã— 4 panels Ã— 3 hours) represented as a 2D array per farm (you can stack/create as shown in tasks).
Farm A panels (rows = panels, cols = hours):


Tasks
Create arrays
Create three 2D NumPy arrays farmA, farmB, farmC (each shape (4,3)).
Stack them into a single 3D array production_all of shape (3, 4, 3).
Array attributes
Print the .shape, .size, .dtype, and .ndim of production_all.
Per-farm totals (arithmetic & aggregation)
Use aggregation ufuncs to compute the total energy per farm (sum across panels & hours) â†’ total_per_farm (shape (3,)).
Compute mean, std, and var per farm.
Per-hour matrix multiplication idea (matrix op)
For a quick weighted estimate, create a hour_weights row vector of shape (3,) = [0.5, 1.0, 0.75] (weights for hour1, hour2, hour3).
Compute weighted_per_panel by performing a dot product between each farm's panel-row and hour_weights to get per-panel weighted energy. (Use np.dot / matrix multiplication appropriately.)
Broadcasting adjustments
The operator notices a fixed calibration offset of +0.5 kWh for all panels in Farm B. Use broadcasting to add this offset to Farm B data inside production_all (without recreating whole array manually).
Element-wise arithmetic (ufuncs: add/sub/mul/div/power/mod)
Compute a normalized version where you:
subtract the per-farm mean (use broadcasting),
divide by the per-farm std (broadcasting) â€” i.e., z-score normalize each farm's panel-hour readings.
Apply many ufuncs (use the ufuncs listed previously)
Apply the following element-wise operations (create variables for each) on the normalized data (or a copy) and show the resulting arrays (or summaries â€” shape/dtype/first row) where appropriate:

Broadcast-based summary table
Use broadcasting to compute a (3,4) array panel_totals_per_farm = total per panel (sum across hours for each panel), and then compute a (3,) boolean needs_maintenance saying True if any panel in that farm has total < 40. Use np.min, np.max, np.any, np.all.
Final KPIs (aggregation)
Compute:
overall_total (sum of all farms),
peak_hour value (max across hours) and which farm & panel produced it (use np.argmax with careful unraveling),
average_per_panel across all farms.

```python
farmA=np.array([[10,12,11],[9,11,10],[12,13,14],[8,7,9]])
farmB=np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farmC=np.array([[15,16,14],[14,13,15],[16,17,18],[12,11,13]])
production_all=np.array([farmA,farmB,farmC])
print(production_all)
print()
print("shape: ",production_all.shape)
print("size: ",production_all.size)
print("Dat type: ",production_all.dtype)
print("Dimensions: ",production_all.ndim)
total_per_farm=np.sum(production_all,axis=(1,2))
print("total per farm: ",total_per_farm)

# Example of elements added for Farm A (production_all[0])
# Summing along axis 1 (panels) and axis 2 (hours)
# For Farm A:
# Panel 1: 10 + 12 + 11 = 33
# Panel 2: 9 + 11 + 10 = 30
# Panel 3: 12 + 13 + 14 = 39
# Panel 4: 8 + 7 + 9 = 24
# Total for Farm A: 33 + 30 + 39 + 24 = 126
# This is the first element in total_per_farm
print("mean: ",np.mean(production_all,axis=(1,2)))
print("std: ",np.std(production_all,axis=(1,2)))
print("var: ",np.var(production_all,axis=(1,2)))
hour_weights=np.array([0.5, 1.0, 0.75])
print()
print("weighted per panel: ",np.dot(production_all,hour_weights))
print("broadcasted 0.5 to b: ",production_all[1]+0.5)
farm_mean=np.mean(production_all,axis=(1,2))
farm_std=np.std(production_all,axis=(1,2))
normalized=(production_all-farm_mean)/farm_std
print("normalized: ",normalized)
print("mean_per_farm: ",farm_mean)
print("std_per_farm: ",farm_std)
print("exp: \n",np.exp(normalized))
print("expm1: \n",np.expm1(normalized))
print("log: \n",np.log(np.clip(normalized,1e-6,None)))
print("log10:\n ",np.log10(np.clip(normalized,1e-6,None)))
print("log2: \n",np.log2(np.clip(normalized,1e-6,None)))
print("sin:\n",np.sin(normalized))
print("cos:\n", np.cos(normalized))
print("tan:\n",np.tan(normalized))
print("arcsin:\n",np.arcsin(np.clip(np.sin(normalized),-1,1)))
print("arccos:\n",np.arccos(np.clip(np.cos(normalized),-1,1)))
print("arctan:\n",np.arctan(normalized))
print("degrees:\n",np.degrees(normalized))
print("radians:\n",np.radians(normalized))
print("round:\n",np.round(normalized))
print("floor:\n",np.floor(normalized))
print("ceil:\n",np.ceil(normalized))
print("trunc:\n",np.trunc(normalized))
high_output = production_all > 20
even_hours_mask=production_all%2==0
print("logival_and: ",np.logical_and(high_output,even_hours_mask))
print("logical_or: ",np.logical_or(high_output,even_hours_mask))
print("logical_not: ",np.logical_not(high_output))
print("logical_xor: ",np.logical_xor(high_output,even_hours_mask))
print("greater_equal: ",np.greater_equal(high_output,even_hours_mask))
print("less: ",np.less(high_output,even_hours_mask))
alerts = np.array([[1,0,1],[0,0,1],[1,1,0],[0,1,0]])
print("bitwise and: ",np.bitwise_and(alerts,high_output))
print("bitwise or: ",np.bitwise_or(alerts,high_output))
print("invert: ",np.invert(alerts))
print("high_output: ",high_output)
print("add: ",np.add(high_output,5))
print("subtract: ",np.subtract(high_output,5))
print("multiply: ",np.multiply(high_output,5))
print("divide: ",np.divide(high_output,5))
print("power: ",np.power(high_output,5))
print("mod: ",np.mod(high_output,5))
print("aggregate sum of np.exp: ",np.sum(np.exp(normalized)))
print("aggregate mean of np.sin: ",np.mean(np.sin(normalized)))
panel_totals_per_farm = np.sum(production_all, axis=(1, 2))
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=0)
needs_maintenance1 = np.min(panel_totals_per_farm < 40, axis=0)
needs_maintenance2 = np.max(panel_totals_per_farm < 40, axis=0)
print("panel_totals_per_farm: ",panel_totals_per_farm)
print("needs_maintenance: ",needs_maintenance)
print("needs_maintenance1: ",needs_maintenance1)
print("needs_maintenance2: ",needs_maintenance2)
overall_total = production_all.sum()
peak_hour_value = np.max(production_all)
peak_farm_panel = np.unravel_index(np.argmax(production_all), production_all.shape)
average_per_panel = np.mean(production_all)
print("Overall total energy:", overall_total)
print("Peak hour value:", peak_hour_value)
print("Farm and panel with peak hour value:", peak_farm_panel)
print("Average per panel across all farms:", average_per_panel)
```
